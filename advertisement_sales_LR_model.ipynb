```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/reklama_sales.csv')
df.head()
```

```python
# Sales ustuni aynan qaysi ustunga ko'proq bog'liqligini aniqlaymiz
df.corrwith(df['Sales'])
```

```python
# Yuqoridagi ga ko'ra, TV ga eng ko'p bo'g'liq ekan
# Endi esa TV ustuni yordamida Simple Linear Reggression model yasaymiz
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='TV', y='Sales', line_kws={'color':'red'})
plt.show()
```

# Simple Linear Reggression

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size=0.1, random_state=42)
```

```python
x_train = np.asanyarray(train_set[['TV']])
y_train = np.asanyarray(train_set[['Sales']])
```

```python
from sklearn.linear_model import LinearRegression
LR_model = LinearRegression()

LR_model.fit(x_train,y_train)
```

**Modelni baholash**

```python
x_test = np.asanyarray(test_set[['TV']])
y_test = np.asanyarray(test_set[['Sales']])
```

```python
y_predict = LR_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

MAE = mean_absolute_error(y_test, y_predict)
RMSE = np.sqrt(mean_squared_error(y_test, y_predict))

print(f"MAE={MAE}")
print(f"RMSE={RMSE}")
```
