```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
url = "https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/diabetes.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
# Avvalo ustunlar orasidagi bog'liqlikni tekshiramiz
df.corrwith(df['Outcome']).sort_values(ascending=False)
```

```python
# Insulin, SkinThickness, BloodPressure larni tashlab yuboramiz, chunki corr juda past
df.drop(['Insulin',"SkinThickness","BloodPressure"], axis=1, inplace=True)
```

```python
df.head()
```

```python
# Endi esa qiymatlarni standardlashtiramiz
X = df.drop('Outcome',axis=1)
y = df['Outcome']

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

# **ML** ni boshlaymiz

```python
# Train/Test set split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=12)
```

```python
# k_NN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=9)

knn.fit(X_train,y_train)
```

```python
y_predict = knn.predict(X_test)
```

# Baholash

Jaccar score

```python
from sklearn.metrics import jaccard_score
jaccard_score(y_test,y_predict)
```

Grid Search


```python
# Endi k ning eng yaxshisini tanlaymiz
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors': np.arange(1, 25)}
knn_gscv = GridSearchCV(knn, param_grid, cv=5)

knn_gscv.fit(X, y)
```

```python
knn_gscv.best_params_ # demak k=24 bo'lganida optimal bo'ladi
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=24)

knn.fit(X_train,y_train)
```

```python
y_predict = knn.predict(X_test)
```

```python
from sklearn.metrics import jaccard_score
jaccard_score(y_test,y_predict)
```
