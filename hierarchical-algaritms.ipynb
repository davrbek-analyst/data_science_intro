```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Avotmobillarni klasterlaymiz
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%204/data/cars_clus.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
# Raqamli ustulnar ham object ko'rinishida bo'lib qolgan. Bularni numeric ga o'tkazamiz
# Nan qiymatlar va hakazolarni tozalaymiz
df[[ 'sales', 'resale', 'type', 'price', 'engine_s',
       'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap',
       'mpg', 'lnsales']] = df[['sales', 'resale', 'type', 'price', 'engine_s',
       'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap',
       'mpg', 'lnsales']].apply(pd.to_numeric, errors='coerce')

df.dropna(inplace=True)
```

```python
# Ba'zi qiymatlarni tashlaganimiz un indexlar o'zgarib ketti, to'g'irlaymiz
df = df.reset_index(drop=True)
df.head()
```

```python
# Klasterlash uchun ustunlarni tanlaymiz
featureset = df[['engine_s',  'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap', 'mpg']]
```

```python
# Normalization
from sklearn.preprocessing import MinMaxScaler
x = featureset.values
max_min_scaler = MinMaxScaler()
featureset_mtx = max_min_scaler.fit_transform(x)
featureset_mtx[:5]
```

# **Klasterlash**
Ierarxik klasterlash uchun Scipy paketidan foydalanmiz.

Boshlanishiga distance matrix D (masofalar matrisasini) quramiz:

```python
import scipy
from scipy import ndimage
from scipy.cluster import hierarchy
from scipy.spatial.distance import cdist
```

```python
leng = featureset_mtx.shape[0]
D = cdist(featureset_mtx, featureset_mtx, 'euclidean')
D
```

# Training the model

```python
Z = hierarchy.linkage(D)
```

```python
fig = plt.figure(figsize=(18,50))
def llf(id):
    return '[%s %s %s]' % (df['manufact'][id], df['model'][id], int(float(df['type'][id])) )

dendro = hierarchy.dendrogram(Z,  leaf_label_func=llf, leaf_rotation=0, leaf_font_size =12, orientation = 'right')
```

```python
# Klaster sonini o'zimiz belgilaymiz
from scipy.cluster.hierarchy import fcluster
k = 5 # 5 ta klaster
clusters = fcluster(Z, k, criterion='maxclust')
clusters
```

**sklearn** yordamida qilish

```python
from sklearn import manifold, datasets
from sklearn.cluster import AgglomerativeClustering
```

```python
agglom = AgglomerativeClustering(n_clusters = 6, linkage = 'complete')
agglom.fit(D)

agglom.labels_
```

# Klasterni taxlil qilamiz


```python
df['cluster'] = agglom.labels_
agg_cars = df.groupby(['cluster','type'])['horsepow','engine_s','mpg','price'].mean()
agg_cars
```
